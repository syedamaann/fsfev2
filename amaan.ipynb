```python
#ROLL-102003620 NAME-SyedAmaan
#Ques 2:
#If the initial and final states are as below and 
#H(n): number of misplaced tiles in the current state n as compared to the goal node need to be considered as the heuristic function. 
#You need to use Best First Search algorithm.
import copy
queueState=[]
visitedState=[]
ctr=0 #variable to store how many times function puzzle is called
class node:
  def __init__(self, node, parent, heuristic):
    self.node=node
    self.parent=parent
    self.heuristic=heuristic
def heuristic(currState, goalState): #function to count number of mismatches and return the count
  ctr=0
  for i in range(3):
    for j in range(3):
      if currState[i][j]!=goalState[i][j]:
        ctr=ctr+1
  return ctr
def indexOfZero(currState): #function to find index of 0
  for i in range(3):
    for j in range(3):
      if currState[i][j]==0:
        return [i, j]
def up(s): #function to move 0 upwards
  currState=copy.deepcopy(s)
  [i, j]=indexOfZero(currState)
  if i>0:
    currState[i][j], currState[i-1][j] = currState[i-1][j], currState[i][j]
  return currState
def down(s): #function to move 0 downwards
  currState=copy.deepcopy(s)
  [i, j]=indexOfZero(currState)
  if i<2:
    currState[i][j], currState[i+1][j] = currState[i+1][j], currState[i][j]
  return currState
def right(s): #function to move 0 towards right
  currState=copy.deepcopy(s)
  [i, j]=indexOfZero(currState)
  if j<2:
    currState[i][j], currState[i][j+1] = currState[i][j+1], currState[i][j]
  return currState
def left(s): #function to move 0 towards left
  currState=copy.deepcopy(s)
  [i, j]=indexOfZero(currState)
  if j>0:
    currState[i][j], currState[i][j-1] = currState[i][j-1], currState[i][j]
  return currState
def notInVisitedState(currState):
  for state in visitedState:
    if currState == state.node:
      return False
  return True
def puzzle(s, g): #iterative function to solve the puzzle
  global ctr
  if s==g:
    return
  new=s
  if notInVisitedState(up(new)) and up(new) != new:
    queueState.append(node(up(new), new, heuristic(up(new), g)))
  if notInVisitedState(down(new)) and down(new) != new:
    queueState.append(node(down(new), new, heuristic(down(new), g)))
  if notInVisitedState(left(new)) and left(new) != new:
    queueState.append(node(left(new), new, heuristic(left(new), g)))
  if notInVisitedState(right(new)) and right(new) != new:
    queueState.append(node(right(new), new, heuristic(right(new), g)))
  queueState.sort(key=lambda node: node.heuristic)
  state=queueState[0]
  visitedState.append(state)
  newState=queueState.pop(0).node
  ctr=ctr+1
  puzzle(newState, g)
path=[]
def pathFinder(visitedNode):
  if visitedNode.parent=="NULL":
    return
  path.append(visitedNode.node)
  for state in visitedState:
    if state.node==visitedNode.parent:
      pathFinder(state)
def main():
  s=[[2,0,3],[1,8,4],[7,6,5]]
  goalState=[[1,2,3],[8,0,4],[7,6,5]]
  queueState.append(node(s, "NULL", heuristic(s, goalState)))
  visitedState.append(node(s, "NULL", heuristic(s, goalState)))
  puzzle(s, goalState)
  print("We've reached the goal state.")
  pathFinder(visitedState[-1])
  path.reverse()
  print("Number of moves:", len(path), "\nPath:", [s]+path)
  print('No. of iterations=', ctr)

if __name__=="__main__":
  main()
```

```python
#ROLL-102003620 NAME-SyedAmaan
#Ques 4:
#If the initial and final states are as below and H(n): Manhattan distance as the heuristic function. You need to use Best First Search algorithm.
import copy
import math

op = ["INITIAL:"]
q = []
visited = []


def find_pos(s):
  for i in range(3):
    for j in range(3):
      if s[i][j]==0:
        return([i,j])
def up(s,pos):
  i=pos[0]
  j=pos[1]
  if i>0: # function to move 0 upwards
    temp = copy.deepcopy(s)
    temp[i][j] = temp[i-1][j]
    temp[i-1][j] = 0
    op.append("UP: ")
    return(temp)
  else:
    return(s)

def down(s,pos): # function to move 0 downwards
    i=pos[0]
    j=pos[1]
    if i<2:
      temp = copy.deepcopy(s)
      temp[i][j] = temp[i+1][j]
      temp[i+1][j] = 0
      op.append("DOWN: ")
      return(temp)
    else:
      return(s)
def right(s,pos): # function to move 0 righwards
  i=pos[0]
  j=pos[1]
  if j<2:
    temp = copy.deepcopy(s)
    temp[i][j] = temp[i][j+1]
    temp[i][j+1] = 0
    op.append("RIGHT: ")
    return(temp)
  else:
    return(s)
def left(s,pos): # function to move 0 leftwards
  i=pos[0]
  j=pos[1]
  if j>0:
    temp = copy.deepcopy(s)
    temp[i][j] = temp[i][j-1]
    temp[i][j-1] = 0
    op.append("LEFT: ")
    return(temp)
  else:
    return(s)

def enqueue(s,value):
  global q
  q = q+[(value,s)]

def dequeue(g):
  global q
  global visited
  q.sort()
  visited = visited+[q[0][1]]
  elem = q[0][1]
  del q[0]
  return(elem)



def index_2d(myList, v):
    for i, x in enumerate(myList):
        if v in x:
            return (i, x.index(v))


def manhatan_dist(board,goal_stat):
    b = board
    g = goal_stat

    manh_dist = 0
    for i in range (3):
        for j in range (3):
            # element value
            bij = b[i][j]

            #element positions
            i_b = i
            j_b = j

            i_g, j_g = index_2d(g,bij)  

            manh_dist += (abs(i_g - i_b) + abs(j_g - j_b))

    return manh_dist

def search(s,g):
  curr_state = copy.deepcopy(s)
  if s == g:
    return
  global visited
  while True:
    pos = find_pos(curr_state)
    new = up(curr_state,pos)
    if new != curr_state:
      if new == g:
        print(op, end=" ")
        print("\nAll the steps are:")
        print(visited+[g])
        return
      else:
        if new not in visited:
          enqueue(new,manhatan_dist(new,g))

    new = down(curr_state,pos)
    if new != curr_state:
      if new == g: 
        print(op, end=" ")
        print("\nAll the steps are:")
        print(visited+[g])
        return
      else:
        if new not in visited:
          enqueue(new,manhatan_dist(new,g))

    new = right(curr_state,pos)
    if new != curr_state:
      if new == g:
        print(op, end=" ")
        print("\nAll the steps are:")
        print(visited+[g])
        return
      else:
        if new not in visited:
          enqueue(new,manhatan_dist(new,g))

    new = left(curr_state,pos)
    if new != curr_state:
      if new == g:
        print(op, end=" ")
        print("\nAll the steps are:")
        print(visited+[g])
        return
      else:
        if new not in visited:
          enqueue(new,manhatan_dist(new,g))

    if len(q)>0:
      curr_state = dequeue(g)
    else:
      print("Not found")
      return

def main():
  s = [[2,0,3],[1,8,4],[7,6,5]]
  g = [[1,2,3],[8,0,4],[7,6,5]]
  global q
  global visited
  q = q
  visited = visited+[s]
  search(s,g)

if __name__ == "__main__":
  main()
```

```python
#ROLL-102003620 NAME-SyedAmaan
#Ques 5: Solve this given problem using Uniform cost search. A is the initial state and G is the goal state
import copy
graph = {
  'S':['A','B','C'],
  'A':['G'],
  'B':['G'],
  'C':['G'],
  'G':[]
}
cost = {
  ('S','A'):1,
  ('S','B'):5,
  ('S','C'):15,
  ('A','G'):10,
  ('B','G'):5,
  ('C','G'):5,
}
totCost=0
path=[]
paths=[]
def problem(node, end):
  global totCost, paths
  if node==end:
    newpath=copy.deepcopy(path)
    paths.append([totCost, newpath])
    path.clear()
    path.append('S')
    totCost=0
    return
  for neighbour in graph[node]:
    totCost+=cost[node, neighbour]
    path.append(neighbour)
    problem(neighbour, end)
  paths.sort()
path.append('S')
problem('S', 'G')
print('The shortest path is: ', end='')
for node in paths[0][1]:
  if node!=paths[0][1][-1]:
    print(node,'->',sep='',end='')
  else:
    print(node,end='')
print(' with tot_cost =', paths[0][0])
```
